```python
import numpy as np
import matplotlib.pyplot as plt

#Parmetros do sinal
fs = 1000 #Frequncia de amostragem original (Hz)
t = np.arange(0, 1, 1/fs)  #Tempo de 0 a 1seg
f1 = 50 #Frequncia do sinal (Hz)
f2 = 120 #Outra frequncia do sinal (Hz)

#Sinal de exemplo
signal = np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)

#Decimao: escolher um fator de decimao
decimation_factor = 4
decimated_signal = signal[::decimation_factor]
decimated_fs = fs/decimation_factor  #Nova frequncia de amostragem

#FFT do sinal decimado
N = len(decimated_signal)
frequencies = np.fft.rfftfreq(N, d = 1/decimated_fs)
fft_values = np.fft.rfft(decimated_signal)

#Plotando os resultados
plt.figure(figsize = (12, 6))

#Sinal original
plt.subplot(2, 1, 1)
plt.plot(t, signal, label = 'Sinal Original')
plt.title('Sinal Original')
plt.xlabel('Tempo(s)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

#FFT do sinal decimado
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(fft_values), label = 'FFT do sinal decimado')
plt.title('FFT do sinal decimado')
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show
```
